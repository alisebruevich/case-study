```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_columns', None)
```

```python
# load data
data = pd.read_csv("alldata.csv")

# cleaning: replace '..' with NaN
for i in data.columns:
    data[i] = data[i].replace({'..': np.NaN})
```

```python
# helper function for finding averages for all developing countries
def country_averages(df):
    averages = {}
    year = 1971

    for i in range(4, len(df.columns)):
        numeric = df.iloc[:,i].astype(float)
        averages[year] = np.mean(numeric)
        year += 1
    
    return averages
```

-------------------------------------

```python
# Birth rate, crude (per 1,000 people)
df_birth_rate = data[data['Series Name'] == "Birth rate, crude (per 1,000 people)"]
```

```python
# plot average life expectancy for all countries over time
averages = country_averages(df_birth_rate)
x = list(averages.keys())
y = list(averages.values())

plt.plot(x,y)
plt.title("Life Expectancy in Developing Countries over Time")
plt.xlabel("Years")
plt.ylabel("Life Expectancy")
plt.show()
```

-------------------------------------

```python
# Cause of death by communicable diseases and maternal, prenatal and nutrition conditions (% total)
df_death_cause_conditions = data[data['Series Name']=="Cause of death, by communicable diseases and maternal, prenatal and nutrition conditions (% of total)"]
```

```python
# plot percent of total deaths caused by communicable diseases and maternal, prenatal and nutrition conditions
averages = country_averages(df_death_cause_conditions)
x = list(averages.keys())
y = list(averages.values())

plt.bar(x,y)
plt.title("Percent of Total Deaths Caused By \n Communicable Diseases and \n Maternal, Prenatal and Nutrition conditions")
plt.xlabel("Years")
plt.ylabel("Percent Deaths")

# for simpler visualization of years
xint = range(1971, 2021, 5)
plt.xticks(xint)
plt.show()
```

-------------------------------------

```python
# Female life expectancy at birth
df_life_exp_female = data[data['Series Name'] == "Life expectancy at birth, female (years)"]
```

```python
# Male life expectancy at birth
df_life_exp_male = data[data['Series Name'] == "Life expectancy at birth, male (years)"]
```

```python
# Total life expectancy at birth
df_life_exp_total = data[data['Series Name'] == "Life expectancy at birth, total (years)"]
```

```python
averages1 = country_averages(df_life_exp_female)
x1 = list(averages1.keys())
y1 = list(averages1.values())

averages2 = country_averages(df_life_exp_male)
x2 = list(averages2.keys())
y2 = list(averages2.values())

averages3 = country_averages(df_life_exp_total)
x3 = list(averages3.keys())
y3 = list(averages3.values())

plt.plot(x1, y1, color="red", label="Female")
plt.plot(x2, y2, color="blue", label="Male")
plt.plot(x3, y3, color="green", label="Total")
plt.legend()
plt.title("Life Expectancy at Birth of Different Groups Over Time")
plt.xlabel("Years")
plt.ylabel("Life Expectancy")
plt.show()
```
