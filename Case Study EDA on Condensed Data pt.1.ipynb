```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_columns', None)
```

```python
# load data
data = pd.read_csv("alldata.csv")

# cleaning: replace '..' with NaN
for i in data.columns:
    data[i] = data[i].replace({'..': np.NaN})
```

```python
# choose 5 least, most developed countries
least_dev_countries = ["Angola", "Benin", "Burkina Faso", "Burundi", "Central African Republic"]
most_dev_countries = ["Mauritius", "Seychelles", "Algeria", "Tunisia", "Botswana"]
```

```python
df_least_dev = data[data['Country Name'].isin(least_dev_countries)]
df_most_dev = data[data['Country Name'].isin(most_dev_countries)]
```

```python
# helper function for finding averages for all developing countries
def country_averages(df):
    averages = {}
    year = 1971

    for i in range(4, len(df.columns)):
        numeric = df.iloc[:,i].astype(float)
        averages[year] = np.mean(numeric)
        year += 1
    
    return averages
```

-------------------------------------

```python
# Birth rate, crude (per 1,000 people)
df_birth_rate_least = df_least_dev[df_least_dev['Series Name'] == "Birth rate, crude (per 1,000 people)"]
df_birth_rate_most = df_most_dev[df_most_dev['Series Name'] == "Birth rate, crude (per 1,000 people)"]
```

```python
averages = country_averages(df_birth_rate_least)
x1 = list(averages.keys())
y1 = list(averages.values())

averages = country_averages(df_birth_rate_most)
x2 = list(averages.keys())
y2 = list(averages.values())

plt.plot(x1,y1, color="blue", label="Least Developed")
plt.plot(x2,y2, color="green", label="Most Developed")
plt.legend()
plt.title("Life Expectancy in Developing African Countries over Time")
plt.xlabel("Years")
plt.ylabel("Life Expectancy")
plt.show()
```
